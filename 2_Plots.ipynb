```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("FullData.csv")
print (data)
```

```python
# plot of stdev x/u* and z/h (v is law of wall estimate of u*)
depth = data['Depth']
u =  3.1925237340085353
std_x_u = np.divide(data['StD_U'],u)
std_x_u_sq = std_x_u*std_x_u
std_z_u = np.divide(data['StD_Z'],u)
std_z_u_sq = std_z_u*std_z_u
turb = data['ta_uwprime']
turb_u = turb/(u*u)
```

```python
nd_depth = np.divide(depth,12.0)
```

```python
# plot of sq stdev u/u* and z/h 

plt.scatter(std_x_u_sq,nd_depth)
plt.xlabel("Normalized Depth")
plt.ylabel("Squared St. Dev. U' / u*")
plt.title('Plot 1')
plt.grid(True)
plt.show()
```

```python
# plot of sq stdev z/u* and z/h

plt.scatter(std_z_u,nd_depth)
plt.xlabel("Normalized Depth")
plt.ylabel("Squared St. Dev. W' / u*")
plt.title('Plot 2')
plt.grid(True)
plt.show()
```

```python
# plot of time average uw / u^2 and z/h 

plot3= plt.scatter(turb_u,nd_depth)
plt.xlabel("Normalized Depth")
plt.ylabel("Average U'W' / u*")
plt.title('Plot 3')
plt.grid(True)
plt.show()
```
